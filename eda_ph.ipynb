```python
def find_household_density(df):
    '''
    Converts addresses dataframe to latitude and longitude by units
    '''
    df['tup_long_lat'] = list(zip(round(df['LATITUDE'],6), round(df['LONGITUDE'],6)))
    agg_df = df['tup_long_lat'].value_counts()
    return agg_df
```

```python
find_household_density(df).to_pickle('long_lat_6_dec.pkl')
```
